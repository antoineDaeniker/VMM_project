```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
from lib.datasets.mpii import MPII
```

```python
# by default, image transform is ImageNormalization

ds = MPII('train')
print(ds[0]['img'].shape, ds[0]['joints'].shape)

ds = MPII('train', transform=None)
print(ds[0]['img'].shape, ds[0]['joints'].shape)
```

```python
ds = MPII('train', transform=None, get_heatmaps=True)
```

```python
for idx in range(100,105):
    sample = ds[idx]
    heatmap = sample['heatmaps']
    meta = sample['meta']
    mask = sample['joints_mask']
    joints = sample['joints']
    
    fig,ax = plt.subplots(1,2,figsize=(10,10))
    ax[0].imshow(sample['img']);
    ax[1].imshow(heatmap.sum(dim=0));

    ax[0].scatter(joints[mask][:,0], joints[mask][:,1], c='red')
    fig.show()
```
